```python
import numpy as np
import pandas as pd
```

```python
students = pd.read_csv("./data/students_info.csv", index_col="STD_ID")
exams = pd.read_csv("./data/exams_info.csv", index_col="STD_ID")
```

```python
students = students[~students.index.duplicated()]
```

```python
students
```

```python
exams
```

```python
passed_students = students.query("Статус == 'ВЫП'").join(exams).reset_index().groupby("STD_ID").agg({"mark": "mean"})
passed_students
```

```python
passed_students.to_csv("./data/output_D.csv")
```
