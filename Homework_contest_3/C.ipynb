```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("./data/exams_info.csv", index_col="STD_ID")
df
```

```python
output = df[df["ДИСЦИПЛИНА"] == "Компьютерные технологии в дизайне"].sort_values(by="ГОД")
output
```

```python
output = output.groupby(["ГОД", "ДИСЦИПЛИНА"]).agg({'mark': 'mean'})
output
```

```python
output.to_csv("./data/output_C.csv")
```
